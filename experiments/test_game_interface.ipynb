## Silent Shitler game interface

This notebook demonstrates usage of the ShitlerEnv environment to play Silent Secret Hitler.

```python
import sys
sys.path.append("../shitler_env")

from game import ShitlerEnv

from pprint import pprint
```

```python
game = ShitlerEnv()
game.reset()

obs = game.observe("P0")
pprint(obs)
```

```python
# Game has one action space
print(game.action_space("P0"))
```

```python
game.step(1)

print(game.observe("P0"))
game.render()
```

```python
for _ in range(5):
    agent_sel = game.agent_selection
    print(f"===== {agent_sel=} =====")
    print("  observation:", game.observe(agent_sel))
    print("  action space:", game.action_space(agent_sel))
    game.step(1)
```

```python
agent_sel = game.agent_selection
pprint(game.observe(agent_sel))
game.render()
print(game.action_space(agent_sel))
```

```python
game.step(0)
agent_sel = game.agent_selection
pprint(game.observe(agent_sel))
print(game.action_space(agent_sel))
```
